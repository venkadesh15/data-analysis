```python
# Dropping the Province/State feature, most of the enteries are null and not important for our annalysis, since we need only country (Nigeria)
hopkins_confirmed.drop('Province/State',axis=1,inplace=True)
hopkins_recovered.drop('Province/State',axis=1,inplace=True)
hopkins_deaths.drop('Province/State',axis=1,inplace=True)
```

```python
hopkins_confirmed.head()
```

```python
hopkins_recovered.head()
```

```python
hopkins_deaths.head()
```

```python
dates=hopkins_confirmed.columns[4:]

h1=hopkins_confirmed.melt(id_vars=['Country/Region','Lat','Long'], value_vars=dates,var_name='Date',value_name='Confirmed')

h2=hopkins_recovered.melt(id_vars=['Country/Region','Lat','Long'],value_vars=dates,var_name='Date',value_name='Recovered')

h3=hopkins_deaths.melt(id_vars=['Country/Region','Lat','Long'],value_vars=dates,var_name='Date',value_name='Deaths')
```

### Extracting Nigeria's Data From John Hopkins Data

```python
h1.head()
```

```python
h2.head()
```

```python
h3.head()
```

```python
# Merging data
d1=pd.merge(h1,h2,on=['Country/Region','Lat','Long','Date'],how='left')
full_data=pd.merge(d1,h3,on=['Country/Region','Lat','Long','Date'],how='left')
```

```python
h1.shape
```

```python
h2.shape
```

```python
h3.shape
```

```python
full_data.shape
```

```python
# Conversion to standard date time feature

full_data['Date']=pd.to_datetime(full_data['Date'])
```

```python
full_data.head()
```

```python
full_data.info()
```

```python
full_data.describe()
```

```python
full_data.isnull().sum()
```

```python
# Creation of another feature called active cases

full_data['Active Cases'] = full_data['Confirmed'] - full_data['Recovered'] - full_data['Deaths']
```

```python
full_data.tail(10)
```

```python
full_data_flat=full_data.set_index(['Country/Region'])
```

```python
# Extracting nigeria cases
nigeria = full_data_flat.loc['Nigeria']
```
